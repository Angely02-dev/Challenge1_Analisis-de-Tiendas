### Grafica

```python
# Función para mostrar porcentaje y cantidad
def autopct_valor(pct, valores):
    total = sum(valores)
    cantidad = int(round(pct * total / 100.0))
    return f"{pct:.1f}%\n({cantidad})"

# Crear figura con 4 subplots (2 filas, 2 columnas)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Lista de datos y títulos
tiendas_data = [
    (categorias_t1, "Tienda 1"),
    (categorias_t2, "Tienda 2"),
    (categorias_t3, "Tienda 3"),
    (categorias_t4, "Tienda 4")
]

# Graficar cada tienda en su subplot
for ax, (categorias, titulo) in zip(axs.flatten(), tiendas_data):
    colores = plt.cm.Paired(range(len(categorias)))
    ax.pie(
        categorias,
        labels=categorias.index,
        autopct=lambda pct: autopct_valor(pct, categorias),
        startangle=90,
        colors=colores,
        wedgeprops={'edgecolor': 'white', 'linewidth': 1},
        explode=[0.05] * len(categorias)
    )
    ax.set_title(f"Cantidad de ventas por Categorias - {titulo}", fontsize=12, fontweight="bold")

plt.tight_layout()
plt.show()
```
